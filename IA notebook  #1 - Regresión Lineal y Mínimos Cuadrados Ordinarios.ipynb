```python
import numpy as np  #Librería y herramientas matemáticas (Multiplicación de matrices, etc.)
import scipy as sc #Tratamiento de imagenes y más herramientas
import pandas as pd #Leer archivos CSV
from sklearn.datasets import load_boston #Dataset importante SKLEARN
from matplotlib import pyplot as plt #Herramienta de visualización de gráficos o datos.
```

```python
#Cargamos la librería de Boston
boston = load_boston()

print(boston.DESCR)
```

```python
x = np.array(boston.data[:, 5]) #Todas las variables de la columna 5. [Dato a escoger, columna]
y = np.array(boston.target) #Con esto vemos el valor medio(MEDV) de cada dato en la columna 5 seleccionada

plt.scatter(x, y, alpha=0.3)
plt.xlabel("Habitaciones")
plt.ylabel("Valor Medio (Miles USD)")

#Se añade una columa de 1s en el vector o matriz, para termino independiente
x = np.array([np.ones(506), x]).T #Se agregan 506 1s, a la matriz, ya que el numero de instancias es 506

b = np.linalg.inv(x.T @ x) @ x.T @ y #Formula MCO, los @ son para multiplicar matrices y np.linalg.inv es para elevar a -1

#b[0] es el valor independiente, b[1] es el corte en el eje Y y se multiplica
#por 4 y 9 ya que son el inicio y el final de la linea roja, 
#el numero de habitaciones mínimas y máximas.
plt.plot([4, 9], [b[0] + b[1] * 4, b[0] + b[1] * 9], c="red") 

plt.show()
```

Formula para minimizar el error cuadrático medio (MCO)
$\beta = (X^{T}X)^{-1}X^{T}Y$

```python
x = np.array(boston.data[:, 5]) 
y = np.array(boston.target) 

plt.scatter(x, y, alpha=0.3)
plt.xlabel("Habitaciones")
plt.ylabel("Valor Medio (Miles USD)")

x = np.array([np.ones(506), x]).T
b = np.linalg.inv(x.T @ x) @ x.T @ y 
plt.plot([4, 9], [b[0] + b[1] * 4, b[0] + b[1] * 9], c="red") 

plt.show()
```

Criminalidad per capita x valor medio de casas en Boston

```python
x = np.array(boston.data[:, 1]) #Todas las variables de la columa 1 - CRIM
y = np.array(boston.target) #Valor medio de las viviendas

plt.scatter(y, x, alpha=0.3)
plt.xlabel("Crimen per capita por pueblo")
plt.ylabel("Valor medio viviendas(Miles USD)")

x = np.array([np.ones(506), x]).T
b = np.linalg.inv(x.T @ x) @ x.T @ y

plt.plot([10, 50], [b[0] + b[1] * 10, b[0] + b[1] * 50], c="red") 
plt.show()
```
